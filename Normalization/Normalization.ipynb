```python
import pandas as pd 
import numpy as np
list_min=[]
list_max=[]
url="phase_signal.csv"
amp=pd.read_csv(url)
amp_df = pd.DataFrame(amp)
#print(amp_df.shape)
list_min=amp_df.min(axis = 0)
list_max=amp_df.max(axis = 0)
min_value=min(list_min)
max_value=max(list_max)
print('max=',max_value)
print('min=',min_value)
dif=max_value-min_value
x_min = min_value
denominator=dif
k =2
z = np.empty(shape=(20921,272))
z.fill(-1)
normalized=(k*(amp_df-x_min)/denominator)+z
print(normalized)
normalized.to_csv('normalized.csv')
print('finished')
```
